```python
dietary_restrictions_count = df['DietaryRestrictions'].value_counts()
dietary_restrictions_count
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
plt.bar(dietary_restrictions_count.index, dietary_restrictions_count.values, color='gold', edgecolor='black')

plt.xticks(dietary_restrictions_count.index, dietary_restrictions_count.index, rotation=90)

plt.xlabel('Dietary Restiction', fontsize=13)
plt.ylabel('Total Count', fontsize=13)

plt.title('Overall Distribution of Bengaluru Restraunts Dietary Restriction by Total Count', fontsize=16)

plt.show()
```

```python
rating_counts = df['rating'].value_counts()
rating_counts
```

```python
colors = ['silver', 'orange', 'yellowgreen', 'red', 'lightpink', 'violet']

plt.figure(figsize=(10,6))
plt.bar(rating_counts.index, rating_counts.values, color=colors[:len(rating_counts)], edgecolor='black')

plt.xticks(rating_counts.index, rating_counts.index, rotation=0)

plt.xlabel('Rating', fontsize=13)
plt.ylabel('Total Number', fontsize=13)

plt.title('Overall Distribution of Bengaluru Restraunts Ratings by Total Number', fontsize=16)

plt.show()
```

```python
top25_cuisines_count = df['cuisine'].value_counts().head(25)
top25_cuisines_count
```

```python
colors = ['gold', 'orange', 'teal', 'mediumaquamarine', 'pink', 'darkcyan']

plt.figure(figsize=(10,6))
plt.bar(top25_cuisines_count.index, top25_cuisines_count.values, color=colors[:len(rating_counts)], edgecolor='black')

plt.xticks(top25_cuisines_count.index, top25_cuisines_count.index, rotation=90)

plt.xlabel('Cuisine', fontsize=13)
plt.ylabel('Total Number', fontsize=13)

plt.title('Overall Distribution of Top 25 Cuisines in Bengaluru by Availability in Numbers', fontsize=16)

plt.show()
```

```python
top30_pincodes_count = df['addressObj/postalcode'].value_counts().head(30)
top30_pincodes_count
```

```python
colors = ['lightpink', 'mediumturquoise', 'deepskyblue', 'lightgreen', 'slategray', 'khaki']

plt.figure(figsize=(10,6))
plt.bar(top30_pincodes_count.index, top30_pincodes_count.values, color=colors[:len(rating_counts)], edgecolor='black')

plt.xticks(top30_pincodes_count.index, top30_pincodes_count.index, rotation=90)

plt.xlabel('Pincode', fontsize=13)
plt.ylabel('Total Number', fontsize=13)

plt.title('Overall Distribution of Top 30 Pincodes in Bengaluru by Density in Numbers', fontsize=16)

plt.show()
```
